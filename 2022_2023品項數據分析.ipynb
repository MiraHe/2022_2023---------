```python
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount("/content/gdrive")
df=pd.read_csv('/content/gdrive/MyDrive/side project/擎雷科技/擎雷.csv',encoding="utf-8")
df = df.fillna(0)
display(df)
```

```python
# 初始化一个字典来映射品项與對應的值
item_mapping = {
    'A': 0,
    'B': 0,
    'C': 0,
    'D': 0,
    'E': 0,
    'F': 0,
    'G': 0,
    'H': 0
}

current_item = None

# 遍歷DataFrame的每一行
for index, row in df.iterrows():
    if row['品項'] in item_mapping:
        current_item = row['品項']
    if current_item:
        df.at[index, '品項'] = current_item

# 顯示结果
display(df)
```

```python
df1 = df.iloc[:, [0] + list(range(2, 11))]

# 打印所選的數據進行驗證
display(df1)
```

```python
# 使用groupby和sum计算每个品项的销售數量總和
sales_total_2023 = df1.groupby('品項').sum().reset_index()
sales_total_2023 = sales_total_2023.astype({'2023年1月': int, '2023年2月': int, '2023年3月': int, '2023年4月': int, '2023年5月': int, '2023年6月': int, '2023年7月': int, '2023年8月': int, '2023年9月': int})

# 打印结果
display(sales_total_2023)
```

```python
# 从数据中选择需要的列（店家代号和品项）
selected_data = df3[['店家代號', '品項']]

# 计算每个店家代号购买每个品项的次数
purchase_counts = selected_data.groupby(['店家代號', '品項']).size().reset_index(name='购买次数')

# 找到前3名店家代号
top_3_shops = purchase_counts.groupby('店家代號')['购买次数'].sum().sort_values(ascending=False).head(3).index

# 在前3名店家代号中找到最常购买的前3名品项
top_items_by_shop = {}
for shop in top_3_shops:
    shop_data = purchase_counts[purchase_counts['店家代號'] == shop]
    top_items = shop_data.sort_values(by='购买次数', ascending=False).head(3)['品項'].tolist()
    top_items_by_shop[shop] = top_items

# 打印结果
for shop, top_items in top_items_by_shop.items():
    print(f"店家代号 {shop} 最常购买的品项： {', '.join(top_items)}")
```

```python
# 2023
# 选择需要的列
selected_data = df[['店家代號', '品項', '2023年1月', '2023年2月', '2023年3月', '2023年4月', '2023年5月', '2023年6月', '2023年7月', '2023年8月', '2023年9月']]

# 创建一个字典以存储前三名店家在每个月的前三名品牌的销售总额
monthly_sales_by_top_shops_and_items = {}

# 计算前三名店家在每个月的前三名品牌的销售总额
for month in selected_data.columns[2:]:
    # 提取每个月的数据
    monthly_data = selected_data[['店家代號', '品項', month]]

    # 找到前三名店家
    top_3_shops = monthly_data.groupby('店家代號').sum().sum(axis=1).sort_values(ascending=False).head(3).index

    # 找到前三名品牌
    top_3_items = monthly_data.groupby('品項').sum().sum(axis=1).sort_values(ascending=False).head(3).index

    # 从月度数据中筛选出前三名店家和前三名品牌的销售数据
    filtered_data = monthly_data[(monthly_data['店家代號'].isin(top_3_shops)) & (monthly_data['品項'].isin(top_3_items))]

    # 计算销售总额
    total_sales = filtered_data.groupby(['店家代號', '品項']).sum().sum(axis=1)

    # 存储在字典中
    monthly_sales_by_top_shops_and_items[month] = total_sales

# 打印结果
for month, sales in monthly_sales_by_top_shops_and_items.items():
    display(f"{month}:")
    display(sales)
```

```python
import seaborn as sns
df4 = pd.DataFrame(monthly_sales_by_top_shops_and_items)

# 创建热图
plt.figure(figsize=(13,10))
sns.heatmap(df4, annot=True, cmap="YlGnBu", fmt='g')
plt.title('2023年前三名店家在每個月的前三名品牌銷售數量')
plt.xlabel('月份')
plt.ylabel('店家代號與品項')
plt.show()
```
